# Progress Review of RL implementation to the problem of STIRAP-like population transfer


## Preliminary work:

Initially I wrote code to implement the STIRAP-like dynamics using QuTiP. My implimentation makes use of the 3 level system hamiltonian 
considered in the following paper https://arxiv.org/pdf/1901.06603.pdf. I then spent some time familiarising myself with openAI's gym as well as their
implimentations package stablebaselines (https://stable-baselines.readthedocs.io/en/master/), which is an imensely useful resource for quick access 
to preprogrammed bleeding edge reinforcement learning agents with working implimentations in most RL algorithms. The stable baselines code relies on duck-typing and works by
interacting with a gym (or 'gym-like') environment (https://gym.openai.com/docs/) and so the next task was to write a custom environment to implement the 
dynamics from my QuTiP code and that had the required structure to be compatable with stable baselines. In the name of generality I wrote a module style python script
that has a ControlEnv base class with the required gym structure and wrote three separate child classes that inherit this structure but have diferent 'action spaces'.
I the wrote a Miscellaneous script containing mostly plotting functions. They can all be found at https://github.com/JBrown742/RL-Stirap-.git.


I have focused mainly on two (rather different) RL algorithms. The first, "Proximal Policy Optimization", is the most recent policy gradient algrithm rolled 
out by openAI and has been seen to perform better on most tasks that their previous algorithms (https://arxiv.org/abs/1707.06347). The second is their implimentation 
of deep Q-learning (https://arxiv.org/abs/1312.5602), which is an all-together different paradigm to traditional policy gradient methods. Where policy gradient methods
interatively improve their policies, deep Q-learning trains deep neural networks learn the optimal quality-function for some MDP, which then in turn informs the actions to take (the policy). 

All trained agents use the stablebaselines default network architectures.

The intention was to look at both algorithms (PPO and DQN) and to impliment them for each of the three different continuous control environment that
I wrote, looking at the different scenarios of resonance, detuning and dephasing as studied in (https://arxiv.org/pdf/1901.06603.pdf). However I have been
unable as yet to obtain any positive results from my ContinuousEnv environement, which employs a continuous action space for both pulses.

 ## Reward Functions:
 
 For each model I have used some form of reward function of the type in https://arxiv.org/pdf/1901.06603.pdf. The reward function had for the PPO section was, for the most part,
 
 \begin{equation}
 r = \alpha (-1 + \rho_{33} - \rho_{22}) - e^{\beta \rho_{22}},
 \end{equation}
 
where the aget tries to bring the first term to zero by maximizing $\rho_{33}$ and minimizing $\rho_{22}$ with an additional punishing term for populating $\rho_{22}$. For all of the PPO models I used $\alpha=1$ and $\beta=3$. In this reward I also included a conditional reward on the population $\rho_{33}$ reaching a certain threshold (0.99). 
        For the DQN implimentation I then assigned the punishing term in the reward function above to a conditional term for when $\rho_{22} > 0.05$.

## Random Unitaries to impliment dephasing:

Here I simply look to write a short code that impliments pure dephasing using random unitary dynamics. We consider a hamiltonian of the form

\begin{equation}
H \propto
\begin{pmatrix} 
0 & \Omega_{P} & 0 \\
\Omega_{P} & \Delta & \Omega_{S} \\
0 & \Omega_{S} & \delta
\end{pmatrix}
\end{equation}

where at each time step we generate several different hamiltonians by randomly sampling $ \Delta$ and $\delta$. we then use these different hamiltonians to evolve the state at the start of that timestep so that we get several final states. We then take a weighted average of these final states as the difinitive final state for that timestep and move on to the next timestep.

   \begin{equation}
   \rho \to \rho_{f} = \sum_{i}^{N} p_i U_i \rho U^{\dagger}_{i}
   \end{equation}
 
where $U_i = e^{-i H_{i} t}$, where $H_i$ are the randomly generated hamiltonians. The code that I have written in attempting to impliment this can be found in the repo linked above https://github.com/JBrown742/RL-Stirap-.
